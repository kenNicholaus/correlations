```python
import os
import numpy as np
import pandas as pd
```

```python
file_path='../data/'
#print(file_path)
date_list = [x[0] for x in os.walk(file_path)]
#print(date_list)
for each_dir in date_list:
        each_file = os.listdir(each_dir)
#print(each_file)
```

```python
with open('../../sp500tickers.txt','r') as f:
    symbol=[line.rstrip('\n') for line in f]
symbol.remove('DX-Y.NYB')
```

```python
#symbol=['^GSPC','^VIX','SPY', 'XLY','XLF','XLK','XLI','XLB','XLV','XLP','XLE','XRT','XLU','^RUT','^DJI']
last_file=pd.to_datetime(each_file[-1].split('.')[0]).date()
for j in symbol[:]:
    print('Started Processing for ', j)
    df=pd.read_csv('../processed/{}.csv'.format(j), index_col=0, parse_dates=True)
    last_date=pd.to_datetime(df.index[-1]).date()
    for i in each_file[0:len(each_file)]:
        iterative_date=pd.to_datetime(i.split('.')[0]).date()
        if iterative_date>last_date:
            print(last_file, last_date, iterative_date)
            df=pd.read_csv(file_path+'/{}'.format(i), index_col=0)
            df.drop(j, axis=1, inplace=True)
            df=df.loc[[j],:]
            df.rename(index={j:iterative_date},inplace=True)
            with open('../processed/{}.csv'.format(j), 'a') as f:
                df.to_csv(f, header=False)
    print('Completed processing of correlations for ', j)
#                 break
```
