```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
from datetime import datetime, timedelta
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
path = '../../prices/'
```

```python
df=pd.read_csv(path+'adjclose.csv', 
               index_col=0, 
               parse_dates=True)
            #.sample(1000, random_state=54)
```

```python
file_path='../data/'
all_dates=[]
date_list = [x[0] for x in os.walk(file_path)]
for each_dir in date_list:
        each_file = os.listdir(each_dir)
last_date=each_file[-1].split('.')[0]
#print(each_file)
for j in each_file:
    all_dates.append(pd.to_datetime(j.split('.')[0]).date())
    #all_dates.append(j.split('.')[0])
#print(all_dates)
```

```python
for i in df.index[::-1]:
    date=pd.to_datetime(i).date()
    if date not in all_dates:
        corr=df[pd.to_datetime(df.index[0]).date():i].corr()
        corr.to_csv('../data/{}.csv'.format(date))
        print('Calculated correlations for: ', date)
    else:
        print('Calculated correlations for all dates')
        break
%run daily_process_correlations.ipynb
```
